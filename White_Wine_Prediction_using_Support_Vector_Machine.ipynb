White Wine Prediction


---



Predict Wine type using Support Vector Machine

Data Set has the following features and targets (columns):-
1.  fixed acidity
2.  volatile acidity
5.  citric acid
6.  residual sugar
7.  chlorides
8.  free sulfur dioxide
9.  total sulfur dioxide
10. density
11. pH
12. sulphates
13. alcohol
14. quality



Data Source is [from here(YBI Foundation GitHub)](https://github.com/YBIFoundation/Dataset/blob/main/WhiteWineQuality.csv)

Import Library

```python
import pandas as pd
```

Import Data

```python
wine = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/WhiteWineQuality.csv', sep = ";")
wine
```

Initial rows of DataFrame

```python
wine.head(17)
```

Information about DataFrame

```python
wine.info()
```

Summary Statistics

```python
wine.describe()
```

Columns of DataFrame

```python
wine.columns
```

Shape of DataFrame

```python
wine.shape
```

Describe / Define data (target(**y**) and features(**X**))

```python
y = wine['quality']
X = wine[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train, X_test, y_train, y_test
```

Modeling

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

Prediction

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

Accuracy test

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
"""That is around 47.45% accuracy score for prediction using SVM for this dataset"""
```
